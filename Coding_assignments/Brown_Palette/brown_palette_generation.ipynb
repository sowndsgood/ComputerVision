# Generation using RGB Format

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

red = np.linspace(80, 130, 5)    # Lower to moderate red values
green = np.linspace(40, 60, 3)   # Lower green values
blue = np.linspace(20, 30, 3) 

plt.figure(figsize=(10,6))

height=5
width=5

rows=5
cols=9

plot_num=1
i=1
for r in red:
    for g in green:
        for b in blue:
            color=np.full((height,width,3),(r,g,b),dtype=np.uint8)
            plt.subplot(rows,cols,plot_num)
            plt.axis('off')
            plt.title(f"Shade{i}")
            plt.imshow(color) 
            plot_num+=1
            i+=1

plt.tight_layout
plt.show()
```

# Generation using HSV Format

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

plt.figure(figsize=(14,10))

h=10
i=1

rows=10
cols=10

height=10
width=10

for s in np.linspace(80,200,10):
    for v in np.linspace(60,140,10):
        color=np.full((height,width,3),(h,s,v),dtype=np.uint8)
        plt.subplot(rows,cols,i)
        color=cv.cvtColor(color,cv.COLOR_HSV2RGB)
        plt.axis('off')
        plt.imshow(color)
        plt.title(f"Shade {i}")
        i+=1

plt.tight_layout()
plt.show()
```
