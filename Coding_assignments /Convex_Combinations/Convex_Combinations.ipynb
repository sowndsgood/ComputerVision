#Problem:

Given two points p1 and p2 how will you generate points on the line using convex combinations?

#Solution:

p new=(λ x1+ (1−λ) x2 , λy1 + (1−λ) y2) where x1,y1 is the first point and x2,y2 is the second point. By varying λ, we can generate new points

```python
p1=(0,0) #Point 1
p2=(3,3) #Point 2
```

```python
x1,y1=p1[0],p1[1] #Take x1 and y1 from point 1
x2,y2=p2[0],p2[1] #Take x2 and y2 from point 2
```

```python
print(x1,x2,y1,y2) #To check
```

```python
import numpy as np  #Importing numpy Library

i=1
points=[]
for q in np.linspace(0,1,15):  #Let q be lambda
  points.append(((q*x1 + (1-q)*x2),(q*y1 + (1-q)*y2),q))
  i+=1
print(points)
```

```python
import matplotlib.pyplot as plt

fig,ax=plt.subplots()
x,y,q=zip(*points)
ax.axhline(y=0,color='black',linewidth=2)
ax.axvline(x=0,color='black',linewidth=2)
ax.plot(x,y,marker='o',color='yellow',label='Line')

ax.plot(x1,y1,marker='o',color='red',label='Point1')
ax.plot(x2,y2,marker='o',color='red',label='Point2')
ax.legend()

for i in range(len(q)):
  ax.text(x[i],y[i],f"λ={q[i]:.2f}")

ax.axis('equal')
ax.grid(True)
plt.show()
```
